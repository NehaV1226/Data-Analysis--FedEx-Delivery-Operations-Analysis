------------

### Handle missing values

```python
# Handle missing values

# There are three columns with missing values, fill missing numerical values with 0 and categorical with 'Unknown'
# Here dosage and shipment mode are categorical and line item insurance is numeric.


# Categorical columns → fill with 'Unknown'
# This would help us find out the missing data entry for this column.
df['Shipment Mode'] = df['Shipment Mode'].fillna('Unknown')
df['Dosage'] = df['Dosage'].fillna('Unknown')



# Numeric column → fill with Mean
# Filling with the mean value would help to create the consistency 
# rather filling with '0' or blindly deleting the column which can create biasness 
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(
    df['Line Item Insurance (USD)'].mean()
)
```

------------

### Convert relevant columns to numeric (if they were read as object)

```python
# Convert relevant columns to numeric (if they were read as object)
cols_to_convert = ['Unit of Measure (Per Pack)', 'Line Item Quantity', 'Line Item Value',
                   'Pack Price', 'Unit Price', 'Weight (Kilograms)', 'Freight Cost (USD)',
                   'Line Item Insurance (USD)']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```

----------

### Remove potential outliers

```python
# Remove potential outliers (optional example: quantities over 500,000)
df = df[df['Line Item Quantity'] <= 500000]
```
